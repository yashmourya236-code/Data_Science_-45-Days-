Polynomial Regression


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split

#data
df = pd.DataFrame({
    "Experience" : [1,2,3,4,5]
})
X = df
y = [10,20,30,90,150]

#poly features
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
print(X_poly)

#model
model = LinearRegression()
model.fit(X_poly, y)

#predict
user_data = int(input("Enter your experience: "))
New_data = pd.DataFrame({
    "Experience" : [user_data]
})
#convert data into poly
u_new_data = poly.transform(New_data)
y_pred = model.predict(u_new_data)
print("The Prediction is : " , y_pred)


#graph
y_poly_pred = model.predict(X_poly)

plt.plot(X, y_poly_pred, color="blue", label="Polynomial Regression")
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.legend()
plt.show()
```

Linear Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier

df = pd.DataFrame({
    "Weight" : [120,140,150,180],
    "Size" : [6,7,7,8]
})

X = df
y = ["Apple" , "Apple" , "Orange" , "Orange"]

#model
model = DecisionTreeClassifier()
model.fit(X,y)
#prediction data
input_weight = int(input("Enter your weight: "))
input_size = int(input("Enter your size: "))
New_data = pd.DataFrame({
    "Weight" : [input_weight],
    "Size" : [input_size]
})

#prediction
predict = model.predict(New_data)
print("Prediction is : " , predict[0])




```

```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier

df = pd.DataFrame({
    "Marks" : [12,40,50,80,100,33] ,
    "Result" : ["Fail" , "Pass" , "Pass" , "Pass" , "Pass" , "Pass"]
})
X = df[["Marks"]]
y = df[["Result"]]

#model
model = DecisionTreeClassifier()
model.fit(X,y)

#prediction data
input_marks = int(input("Enter your marks: "))
New_data = pd.DataFrame({
    "Marks" : [input_marks]
})

#prediction
predict = model.predict(New_data)
print(" Prediction is : " , predict[0])
```

Logistic Regression

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression


X = pd.DataFrame({
    "Hours": [1, 2, 3, 4, 5, 6]
})

y = [0, 0, 0, 1, 1, 1]

model = LogisticRegression()

model.fit(X, y)
# new data
data = int(input("Enter your data: "))
new_data = pd.DataFrame({
    "Hours":[data]
})

print(model.predict(new_data))
```
