```python
import pandas as pd
d = {
    "name":["vishal","vineet","virat"],      #3rows
    "salary":[100,200,300]                  #3rows
}
df=pd.DataFrame(data=d)
df["holidays"] = df["salary"] / 100
df["increment"] = [10,20,30]
#delete column
df.drop(["salary","name"],axis=1)
df
```
