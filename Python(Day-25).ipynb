```python
#Linear Regression
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

Data = {
    "Experience" : [1,2,3,4,5] ,
    "Salary" : [3,5,7,9,11]
}
df = pd.DataFrame(Data)
print("                   DATA                   ")
print(df)
print('\n')

#data split
X = df[['Experience']]
Y = df['Salary']

#model
model = LinearRegression()
model.fit(X,Y)

print("__________________________________________________________")
print('\n')
#prediction
exp = int(input("Enter Your Experience : "))
new_data = pd.DataFrame({
    "Experience":[exp]
})

pred = model.predict(new_data)
print("The Predicted Salary Is : " , "$",pred[0])
```

```python
# evaluation metrics
import numpy as np
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error , r2_score

# actual values prediction
y_true = np.array([50,60,75,90,100])

# predicted values
y_pred = np.array([52,58,78,88,105])

# mean absolute error
mae = mean_absolute_error(y_true,y_pred)
print(mae)

#mse (mean squared error)
mse = mean_squared_error(y_true,y_pred)
print(mse)

# r2 score
r2 = r2_score(y_true,y_pred)
print(r2)

#rmse
rmse = np.sqrt(mse)
print(rmse)

```

```python
# lasso vs ridge expression

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge,Lasso
from sklearn.metrics import median_absolute_error

data = {
    "size":[1000,1200,1500,1800,2000,2200,2500,2700,3000,3200],
    "bedrooms":[2,2,3,3,3,4,4,4,5,5],
    "age":[20,18,15,12,10,8,6,5,4,2],
    "price":[30,35,45,50,55,60,68,72,80,85]
}

df = pd.DataFrame(data)
print(df)

# split data in x or y
x = df[['size','bedrooms','age']]
y = df['price']

# splitting data between training or testing
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)

#   training -> ridge regression
ridge = Ridge(alpha=1.0)
ridge.fit(x_train,y_train)

# prediction
ridge_p = ridge.predict(x_test)
print("x test")
print(x_test)
print("prediction")
print(ridge_p)
```

```python
# logistic regression
import pandas as pd
from sklearn.linear_model import LogisticRegression
data = {
    "hours":[1,2,3,4,5,6,7],
    "result":[0,0,0,1,1,1,1]
  }

df = pd.DataFrame(data)
print(df)

# split x and y
x = df[['hours']]
y = df['result']

# model
model = LogisticRegression()
model.fit(x,y)

# prediction
new_data = pd.DataFrame({
    "hours":[5.5]
})
pred_d = model.predict(new_data)
print(pred_d)

# prediction of probability
prob_d = model.predict_proba(new_data)
print(prob_d)
#
```

```python

import pandas as pd
from sklearn.neighbors import KNeighborsClassifier

#dataset
data = {
    "weight":[100,110,120,170,180,190],
    "fruits":[0,0,0,1,1,1] # 0 -> apple and 1 -> orange
}

df = pd.DataFrame(data)
print("                   -DATA-                   ")
print(df)
print('\n')
print("__________________________________________________________")
print('\n')

#split
X = df[["weight"]]
y = df["fruits"]

#model
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X,y)

#prediction
weight = int(input("Enter Weight : "))
new_data = pd.DataFrame({
    "weight" : [weight]
})

pred = model.predict(new_data)
print("The Predicted Fruit Is : ",pred)
```
