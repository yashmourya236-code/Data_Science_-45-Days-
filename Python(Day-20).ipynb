```python
# handle missing values
import pandas as pd

data = {
    'Name': ['Ram', 'Kamal', 'Ajay'],
    'Age': [25, None, 32],
    'Salary': [5000, 6000, None]
}

df = pd.DataFrame(data)

print("Original Dataframe")
print(df)

# Find null values
print(df.isnull().sum())

# Drop rows with null values
df_drop = df.dropna()
print(df_drop)

# Fill missing values with mean
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())

print(df)
```

```python
# handle missing values
import pandas as pd
import numpy as np
data = {
"colors":['red','green','blue','orange','green','blue',np.nan]
}
df = pd.DataFrame(data)
# print(df)

# handle null values
df.dropna(inplace=True)
print(df)
```

```python
# label encoding
from sklearn.preprocessing import LabelEncoder

# method 1
le = LabelEncoder()
df['colors_encoder1'] = le.fit_transform(df['colors'])

# method 2
df['colors_encoder2'] = LabelEncoder().fit_transform(df['colors'])
# print(df)

# method 3
import sklearn
df['colors_encoder3'] = sklearn.preprocessing.LabelEncoder().fit_transform(df['colors'])
print(df)
```

```python
# Combination of -> H.m.v | L.E | O.H.E

import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder , OneHotEncoder
data = {
    "age":[10,20,np.nan,26,30],
    "gender":['male','female','others','male',np.nan],
    "name":['raj','priya','yash','hello','veer']
}

df = pd.DataFrame(data)
print(df)

# Handle missing values
df['age'] = df['age'].fillna(df['age'].mean())
df.dropna(subset=['gender'],inplace=True)
print(df)

#label Encoding
le= LabelEncoder()
df['gender']= le.fit_transform(df['gender'])
print(df)

# one hot encoder
oe= OneHotEncoder()
oe_e= oe.fit_transform(df[['gender']]).toarray()
print(oe_e)
```
