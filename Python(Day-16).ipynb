```python
# creating first chart of line graph
import matplotlib.pyplot as plt
x = [2010,2015,2020,2025]
y = [100,200,250,300]
plt.plot(x,y)
plt.show()
```

```python
# creating | mentioning the "year" , "sales" , "sales growth"
import matplotlib.pyplot as plt
x = [2010,2015,2020,2025]
y = [100,200,250,300]
plt.plot(x,y)
plt.xlabel("year")
plt.ylabel("sales")
plt.title("sales report")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [2010,2015,2020,2025]
y = [100,200,250,300]
# graph size
plt.figure(figsize=(10,4)) # width or height | x or y
plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [2010,2015,2020,2025]
y = [100,200,250,300]
# graph size
plt.figure(figsize=(10,4)) # width or height | x or y
plt.plot(x,y,color="blue",marker='o',linestyle="dashed",linewidth = 2,markersize=10)
plt.show()
```

```python
#multiple line chart
import matplotlib.pyplot as plt
x = [2010,2015,2020,2025]
y1 = [100,200,250,300]
y2 = [150,105,195,300]
plt.plot(x,y1,label="jeans")
plt.plot(x,y2,label="shirt")
plt.legend()# info of label
plt.show()
```

```python
#multiple line chart
import matplotlib.pyplot as plt
x = [2010,2015,2020,2025]
y1 = [100,200,250,300]
y2 = [20,49,78,99]
plt.xlabel("year")
plt.ylabel("population")
plt.title("populations per year")
plt.figure(figsize=(6,4)) # width or height | x or y
plt.plot(x,y1,color="blue",marker='o', label = "india",linestyle="dashed",linewidth = 2,markersize=1)
plt.plot(x,y2,color="yellow",label = "uk",marker='o',linestyle="dashed",linewidth = 2,markersize=1)
# plt.plot(x,label="india")
# plt.plot(y1,label="uk")
plt.legend()# info of label
plt.show()
```

```python
import matplotlib.pyplot as plt
x = ["group a ","group b","group c","group d"]
y1 = [10,20,20,40]
y2 = [20,30,25,30]
plt.bar(x,y1,label = "jeans",color="blue")
plt.bar(x,y2,label="shirt",color="violet")
plt.legend()
plt.xlabel("group")
plt.ylabel("no. of students")
plt.title("no.of students in each group")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3,4])
y1=[10,20,20,40]
y2=[20,30,25,30]
# calculation ->width
w = 0.40
plt.bar(x - w/2,y1 , label="boys",width=w)# hide second
plt.bar(x + w/2,y2 , label="girls",width=w)# show second
plt.legend()
plt.xlabel("group")
plt.ylabel("no. of students")
plt.title("no.of students in each group")
plt.show()
```
