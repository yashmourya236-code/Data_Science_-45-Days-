```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3,4])
y1=[10,20,30,40]
y2=[20,30,55,90]
y3=[10,30,70,90]
# calculation ->width
w = 0.30
plt.bar(x - w,y1 , label="boys",width=w)# hide second
plt.bar(x ,y2 , label="girls",width=w)# show second
plt.bar(x + w,y3 , label="boys and girls",width=w)
plt.legend()
plt.xlabel("boys")
plt.ylabel("girls")
plt.title("no.of students")
plt.show()
```

```python
# HISTOGRAM CHART
import matplotlib.pyplot as plt
marks = [40,55,60,70,75,90,33,50]
plt.hist(marks,bins=2,color="violet")
plt.xlabel("MARKS")
plt.ylabel("NO.OF STUDENTS")
plt.title("HISTOGRAM")
plt.show()
```
